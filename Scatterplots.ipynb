# Scatterplots

### Gráficos de Dispersão

Dataset:<br>
https://www.kaggle.com/shwetabh123/mall-customers

```python
import pandas as pd
```

```python
df = pd.read_csv('datasets/mall_customers.csv')
```

```python
df.plot.scatter(x = 'Age',
               y = 'Annual Income (k$)',
               c = 'Darkblue',
               figsize = (10, 5));
```

```python
df.plot.scatter(x = 'Age',
               y = 'Annual Income (k$)',
               c = 'Spending Score (1-100)',
               colormap = 'viridis',
               figsize = (10, 5));
```
