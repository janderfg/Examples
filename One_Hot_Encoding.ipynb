# One Hot Encoding

```python
import pandas as pd
import numpy as np
```

```python
# Criando um dataset de exemplo
df = pd.DataFrame({'Nome': ['Dayana', 'Camila', 'Felipe', 'Dayana', 'Jander', \
                            'Paulo', 'Jander', 'Marcos', 'Oswaldo', 'Gabriela'],
                   'Cidade': ['Mogi-Mirim', 'Campinas', 'São Paulo', 'Campinas', 'Mogi-Mirim', \
                              'Campinas', 'Goiânia', 'Itapira', 'Limeira', 'Mogi-Mirim'],
                   'Idade': [34, 25, 17, 28, 43, 24, 15, 28, 15, 27], 
                   'Sexo': ['F', 'F', 'M', 'F', 'M', 'M', 'M', 'M', 'M', 'F'],
                   'Escolaridade': ['Ensino Fundamental', 'Ensino Médio', 'Ensino Médio', 'Ensino Superior', 'EnsinoFundamental', \
                                   'Gradução', 'Pós-Gradução', 'Ensino Médio', 'Ensino Superior', 'Ensino Médio']})
```

```python
df
```

### One Hot Encoding

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
ohe = OneHotEncoder()
```

```python
x = df['Cidade'].values
```

```python
x = ohe.fit_transform(x.reshape(-1,1)).toarray()
```

```python
x
```

```python
df = np.concatenate([df, x], axis=1)
```

```python
df = pd.DataFrame(df)
```

```python
df
```

```python
df.rename(columns={0: 'Nome',1:'Cidade',2:'Idade',3:'Sexo',4:'Escolaridade',5:'Campinas',\
                   6:'Goiânia',7:'Itapira',8:'Limeira',9:'Mogi-Mirim',10:'São Paulo'}, errors='raise')
```
