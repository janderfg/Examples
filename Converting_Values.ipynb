# Convertendo valores

```python
import pandas as pd
```

```python
df = pd.DataFrame({'Nome': ['Lucas', 'Pedro', 'Maria', 'Paulo', 'Eliana'],
                   'Sexo': ['masculino', 'masculino', 'feminino', 'masculino', 'feminino'],
                   'Fumante': ['não', 'não', 'sim', 'não', 'sim'],
                   'Idade': [16, 19, 23, 17, 32],
                   'Mora sozinho?': ['sim','não','não','sim','não'],
                   'Brasileiro?': ['não','sim','sim','sim','sim'],
                   'Quantos Filhos?': ['nenhum','dois','nenhum','nenhum','um']})
```

```python
df
```

```python
# Map
df['Sexo_MF'] = df['Sexo'].map({'masculino': 'M', 'feminino': 'F'})

# Factorize
df['Fumante_01'] = df['Fumante'].factorize()[0]

# Booleano
df['Idade_para_dirigir'] = df['Idade'] >= 18

# List Comprehension
df['Mora_Sozinho'] = [0 if x == 'sim' else 1 for x in df['Mora sozinho?']]

# Apply
df['Brasileiro'] = df['Brasileiro?'].apply(lambda x: 0 if x == 'não' else 1)

# Substituição
df.loc[df['Quantos Filhos?'] == 'nenhum', 'Qtde_Filhos'] = 0
df.loc[df['Quantos Filhos?'] == 'dois', 'Qtde_Filhos'] = 2
df.loc[df['Quantos Filhos?'] == 'um', 'Qtde_Filhos'] = 1
```
